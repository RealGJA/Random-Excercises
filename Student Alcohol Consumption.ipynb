```python
import pandas as pd
import numpy

df = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv')
df.head()
```

```python
stud_alcohol = df.loc[:,'school':'guardian']
stud_alcohol
```

```python
capitalizer = lambda x: x.capitalize()
```

```python
stud_alcohol['Mjob'].apply(capitalizer)
stud_alcohol['Fjob'].apply(capitalizer)
```

```python
stud_alcohol.tail()
```

```python
stud_alcohol['Mjob'] = stud_alcohol['Mjob'].apply(capitalizer)
stud_alcohol['Fjob'] = stud_alcohol['Fjob'].apply(capitalizer)
```

```python
stud_alcohol.tail()
```

```python
def majority(x):
        if x > 17:
            return True
        else:
            return False

stud_alcohol['legal_drinker'] = stud_alcohol['age'].apply(majority)
```
